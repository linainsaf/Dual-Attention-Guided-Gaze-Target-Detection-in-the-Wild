```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import os
import pandas as pd
import pickle
import glob
from PIL import Image
import math
from skimage.transform import resize
```

```python
def load_images_from_folder(folder):
    depth_maps = []
    depth_maps_name = []    
    for filename in os.listdir(folder):
        img = cv.imread(os.path.join(folder,filename))
        if img is not None:
            depth_maps.append(img) 
            depth_maps_name.append(filename)
    return depth_maps, depth_maps_name
```

```python
depth_maps = []
depth_maps_name = []
depth_maps, depth_maps_name = load_images_from_folder("C:/Users/Admin/Documents/UPMC_ISI_2/ML/projet/Dual-Attention-Guided-Gaze-Target-Detection-in-the-Wild/depth_estimation/output")
```

```python
depth_maps=np.reshape(depth_maps,(4000,2))
depth_maps_name=np.reshape(depth_maps_name,(4000,2))
```
